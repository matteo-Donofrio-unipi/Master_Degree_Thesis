```python
import pandas as pd
abstract_df = pd.DataFrame(columns=['Paper','From','Date','Title','Authors'])
abstract_df
```

```python
keys = ['Paper','Title','Authors']

for i in ['1992', '1993', '1994', '1995', '1996', '1997', '1998', '1999']:
    os.chdir(cwd+ '\\'+i+'\\')
    print(os.getcwd())
    #print(os.listdir(os.getcwd()))
    
    paper_list = os.listdir(os.getcwd())
    
    for j in paper_list:
    
        abstracts_temp = pd.read_csv(j, sep = "\t")
        
        col_name = abstracts_temp.columns
        abstracts_temp.rename(columns={col_name[0]: "rows"}, inplace = True)
        
        

        for k in range(len(abstracts_temp)):
            if(abstracts_temp.iloc[k]['rows'].startswith('Paper')):
                Paper = abstracts_temp.iloc[k]['rows']
            elif(abstracts_temp.iloc[k]['rows'].startswith('Title')):
                Title = abstracts_temp.iloc[k]['rows']
            elif(abstracts_temp.iloc[k]['rows'].startswith('Authors')):
                Authors = abstracts_temp.iloc[k]['rows']
        
        serie = pd.Series({'Paper' : Paper, 'Title':Title,'Authors':Authors})
        abstract_df = pd.concat([abstract_df, serie.to_frame().T])
        
    print("###########################################")
    

abstract_df
```
